Visualisasi

```python
dataPenjualan = {'Donat': 14, 'Pastel': 5,'Kue Coklat': 12,'Ice Cream': 19,'Puding Vanila': 6,'Brownies': 8,'Puding Strawberi': 12,'Puding Coklat': 9,'Ice Cream Nutela': 10,'Kue Coklat-Keju': 17,'Kue Wajik': 2,'Kue Sus': 9,'Mochi': 13}
df_dataPenjualan = pd.DataFrame(list(dataPenjualan.items())).rename(columns={0:"Nama Makanan", 1:"Hasil Penjualan"})
df_dataPenjualan.sort_values(by="Hasil Penjualan", inplace=True)
x_coor = np.arange(len(df_dataPenjualan))

colors = []
for _ in range(len(df_dataPenjualan)) :
    colors.append('#FF1493' if _ < 3 else '#00FFFF')

plt.barh(x_coor,df_dataPenjualan["Hasil Penjualan"], tick_label=df_dataPenjualan["Nama Makanan"], color=colors)
plt.ylabel("Hasil Penjualan")
plt.title("Penjualan Kantin Biru")
plt.show()
```

3 makanan penutup yang harus disingkirkan adalah : Puding vanila, pastel, dan kue wajik
